### Import all the csv files into Pandas dataframes.

```python
path = r'/Users/jin/link/credit/homecredit_project/home-credit-default-risk'
all_files = glob.glob(os.path.join(path, "*.csv"))

# There are 10 csv files.
homecredit_df = {}
for file in all_files:
    file_name = os.path.basename(file)[:-4]
    homecredit_df[file_name] = pd.read_csv(file, index_col=None, header=0, encoding="ISO-8859-1")    

# Among the 10 csv files, 2 are just for reference purpose. I seperate them from the other 8 here.
cols_description = homecredit_df['HomeCredit_columns_description']
cols_description = cols_description.drop(cols_description.columns[0], axis=1)
sample_submission = homecredit_df['sample_submission']
del homecredit_df['HomeCredit_columns_description']
del homecredit_df['sample_submission']
```

### Import all the Pandas dataframes into MySQL tables.

```python
# First I have to check what data types are in each dataframe.
for df in homecredit_df.keys():
    print(df)
    print(homecredit_df[df].dtypes.value_counts())
    print()
    
# It seems like we only have 3 data types: float64, int64, object.
```

```python
# Write a SQL query to create MySQL table schemas for each df.
def create_table(df, df_name, mycursor):
    cols_query = f"CREATE TABLE {str(df_name)} ("
    
    # Assigin appropriate dtypes to each column.
    for col in df.columns:
        if df[col].dtype == 'O':
            cols_query = cols_query + col + " text, "
        elif df[col].dtype == 'int64':
            cols_query = cols_query + col + " int, "
        else:
            cols_query = cols_query + col + " float, "
    cols_query = cols_query[: -2] + ")"
                                                  
    mycursor.execute(cols_query)
    print(f'table {str(df_name)} is created.')
    
# Add rows to each mySQL table generated above.
# Since the dataframes are huge, this process is done in batches.
def Add_rows(df, df_name, mycursor, batch_size=500000):
    rows_query = f"INSERT INTO {str(df_name)} VALUES (" + '%s, ' * len(df.columns)
    rows_query = rows_query[: -2] + ")"
    
    # Make all the columns object type in order to replace nan to None even for numerical columns.
    df = df.astype('O')
    
    n_iters = (len(df) // batch_size) + 1
    print(f'{n_iters} batches')
    
    t0 = time()
    for i in range(n_iters):
        batch_df = df.iloc[i*batch_size:(i+1)*batch_size, :]
        batch_df = batch_df.where(pd.notnull(batch_df), None)
        batch = list(batch_df.itertuples(index=False, name=None))
        
        mycursor.executemany(rows_query, batch)
        print(f'batch {i} is added to {df_name}')
        
    time_consumed = time() - t0
    print(f'all rows are added to table {str(df_name)} and it took {time_consumed} seconds.')
    print()

mydb = mysql.connector.connect(
host="localhost",
user="root",
passwd="passwd1234",
database="homecredit")

mycursor = mydb.cursor()

for df_name in homecredit_df.keys():
    create_table(homecredit_df[df_name], df_name, mycursor=mycursor)
    Add_rows(homecredit_df[df_name], df_name, mycursor=mycursor)

mydb.commit()
```
